Bạn hãy tạo một mẫu gồm 10000 phần tử mô phỏng chiều cao của nam thanh niên có giá trị từ 120cm - 200cm. Bạn hãy lưu kết quả vào biến **POP**.

Tính:

b. Tính tỷ lệ người cao trong quần thể, biết rằng thanh niên có chiều cao từ 180 trở lên được gọi là cao.

c. Vẽ histogram về chiều cao của quần thể. Theo bạn quần thể có phân phối chuẩn hay không?

d. Thử vẽ histogram và đồ thị hàm mật độ của phân phối chuẩn sử dụng tham số loc và scale bằng với kỳ vọng và độ lệch chuẩn của quần thể.

```python
pop = np.random.randint(120,200, size=10000)
print(pop[:10])
```

```python

mean_height = np.mean(pop)
print('Chieu cao tung binhf cuar quan the:',mean_height)
std_height = np.std(pop)
print('do lech chuan ve chieu cao cua quan the la:',std_height)
```

```python
tall_people = sum(pop>=180)/len(pop)
tall_ratio = tall_people/10000
print('Ty le nguoi cao trong quan the la:',tall_people)
```

```python
mu,std = norm.fit(pop)
xmin,xmax = plt.xlim()
x = np.linspace(xmin,xmax,100)
p = norm.pdf(x,mu,std)
plt.plot(x,p,'k',linewidth=2)
plt.xlabel('chieu cao')
plt.ylabel('tan so')
plt.title('phan phoi chieu cao')
plt.show
```

```python
plt.hist(pop,bins=50)
plt.xlabel('chieu cao')
plt.ylabel('tan suat')
plt.title('Phan phoi chieu cao cuar quan the')
plt.show
```
